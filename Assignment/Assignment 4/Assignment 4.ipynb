```python
%config InteractiveShell.ast_node_interactivity='all'
import numpy as np
np.set_printoptions(threshold=np.inf)

# Construct the matrix A
import numpy as np
n=40
A=np.zeros((n,n))
A+=np.diag(-1*np.ones(39),-1)+np.diag(2*np.ones(40))+np.diag(-1*np.ones(39),1)
```

```python
# Compute the eigenvectors and eigenvalues
eigenvalues, eigenvectors=np.linalg.eig(A)
# Sort the eigenvalues and eigenvectors by eigenvalue in ascending order
sort_indices = np.argsort(eigenvalues)
eigenvalues = eigenvalues[sort_indices]
eigenvectors = eigenvectors[:,sort_indices]
print('Here is all the eigenvalues:',eigenvalues)
eigen_norm3rd_for_plot=[]
for i in range(len(eigenvectors)):
    eigen_norm3rd_for_plot.append(-eigenvectors[i][2])
eigen_norm3rd_for_plot=eigen_norm3rd_for_plot/np.linalg.norm(eigen_norm3rd_for_plot,ord=np.inf)
eigen_norm6th_for_plot=[]
for i in range(len(eigenvectors)):
    eigen_norm6th_for_plot.append(eigenvectors[i][5])
eigen_norm6th_for_plot=eigen_norm6th_for_plot/np.linalg.norm(eigen_norm6th_for_plot,ord=np.inf)
eigen_norm10th_for_plot=[]
for i in range(len(eigenvectors)):
    eigen_norm10th_for_plot.append(-eigenvectors[i][9])
eigen_norm10th_for_plot=eigen_norm10th_for_plot/np.linalg.norm(eigen_norm10th_for_plot,ord=np.inf)
# # Normalise the eigenvector in 2-norm
# eigenvectors_2norm_val=np.linalg.norm(eigenvectors,ord=2,axis=1)
# eigenvectors_2norm=eigenvectors/eigenvectors_2norm_val

# x point
x_n=np.linspace(0,1,42)
# For the 3rd smallest eigenvalue
eigen_norm3rd_for_plot=np.append(eigen_norm3rd_for_plot,0)
eigen_norm3rd_for_plot=np.insert(eigen_norm3rd_for_plot,0,0)
import matplotlib.pyplot as plt 
plt.figure(figsize=(15,5),dpi=300)
plt.subplot(1,3,1)
plt.title('3rd eigenvalue')
plt.plot(x_n,eigen_norm3rd_for_plot)
plt.xlabel('x_n')
plt.ylabel('y')
# For the 6th smallest eigenvalue
eigen_norm6th_for_plot=np.append(eigen_norm6th_for_plot,0)
eigen_norm6th_for_plot=np.insert(eigen_norm6th_for_plot,0,0)
plt.subplot(1,3,2)
plt.title('6th eigenvalue')
plt.plot(x_n,eigen_norm6th_for_plot)
# For the 10th smallest eigenvalue
eigen_norm10th_for_plot=np.append(eigen_norm10th_for_plot,0)
eigen_norm10th_for_plot=np.insert(eigen_norm10th_for_plot,0,0)
plt.xlabel('x_n')
plt.ylabel('y')
plt.subplot(1,3,3)
plt.title('10th eigenvalue')
plt.plot(x_n,eigen_norm10th_for_plot)
plt.xlabel('x_n')
plt.ylabel('y')
plt.savefig('eigenvector of A')
```
