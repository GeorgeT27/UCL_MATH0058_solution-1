```python
# Construct the matrix A
import numpy as np
n=40
A=np.zeros((n,n))
A+=np.diag(-1*np.ones(39),-1)+np.diag(2*np.ones(40))+np.diag(-1*np.ones(39),1)
A
```

```python
# Compute the eigenvectors and eigenvalues
eigenvalues, eigenvectors=np.linalg.eigh(A)
# Sort the eigenvalues and eigenvectors by eigenvalue in ascending order
sort_indices = np.argsort(eigenvalues)
eigenvalues = eigenvalues[sort_indices]
eigenvectors = eigenvectors[:,sort_indices]
# Normalise the eigenvector
eigen_norm=[]
for eigenvector in eigenvectors:
    sup_vector=np.max(np.abs(eigenvector))
    eigenvector_norm=eigenvector/sup_vector
    eigen_norm.append(eigenvector_norm)
    
# # Normalise the eigenvector in 2-norm
# eigenvectors_2norm_val=np.linalg.norm(eigenvectors,ord=2,axis=1)
# eigenvectors_2norm=eigenvectors/eigenvectors_2norm_val
```

```python
# x point
x_n=np.linspace(0,1,42)
# For the 3rd smallest eigenvalue
eigen_norm3rd_for_plot=np.append(eigen_norm[2],0)
eigen_norm3rd_for_plot=np.insert(eigen_norm3rd_for_plot,0,0)
import matplotlib.pyplot as plt 
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.title('3rd eigenvalue')
plt.plot(x_n,eigen_norm3rd_for_plot)
plt.xlabel('x_n')
plt.ylabel('y')
# For the 6th smallest eigenvalue
eigen_norm6th_for_plot=np.append(eigen_norm[5],0)
eigen_norm6th_for_plot=np.insert(eigen_norm6th_for_plot,0,0)
plt.subplot(1,3,2)
plt.title('6th eigenvalue')
plt.plot(x_n,eigen_norm6th_for_plot)
# For the 10th smallest eigenvalue
eigen_norm10th_for_plot=np.append(eigen_norm[9],0)
eigen_norm10th_for_plot=np.insert(eigen_norm10th_for_plot,0,0)
plt.xlabel('x_n')
plt.ylabel('y')
plt.subplot(1,3,3)
plt.title('10th eigenvalue')
plt.plot(x_n,eigen_norm10th_for_plot)
plt.xlabel('x_n')
plt.ylabel('y')
```

```python
#Compare with the acutal eigenvectors
x=np.linspace(0,1,1001)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(x,np.sin(3*np.pi*x))
plt.plot(x_n,eigen_norm3rd_for_plot)
plt.subplot(1,3,2)
plt.plot(x,np.sin(6*np.pi*x))
plt.plot(x_n,eigen_norm6th_for_plot)
plt.subplot(1,3,3)
plt.plot(x,np.sin(10*np.pi*x))
plt.plot(x_n,eigen_norm10th_for_plot)
```
